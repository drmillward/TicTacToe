```python
def marker_assign():
    global player1
    global player2
    global active_player
    global turn
    global game_status
    # asks if the player 1 would like to be X or O and starts the game
    #print('Welcome to tictactoe!!!')
    x = str(input('would player 1 like to be X or O? '))
    if x.upper() == 'X':
        player1 = player_icon[0]
        player2 = player_icon[1]
        active_player = ['player 1', player1]
        print('Player 1 is X')
        turn = 0
        prtbrd(board)
        game_status = 1
    elif x.upper() == 'O':
        player1 = player_icon[1]
        player2 = player_icon[0]
        active_player = ['player 1', player1]
        print('player 1 is O')
        turn = 0
        prtbrd(board)
        game_status = 1
    # if input is invalid, runs the function again   
    else:
        print('please input X or O')
        marker_assign()
```

```python
def win_check(board):
    global active_player
    global game_status
    
    #checks the win conditions
    if (board[8] == board[7] == board[6] or 
    board[5] == board[4] == board[3] or 
    board[2] == board[1] == board[0] or 
    board[8] == board[5] == board[2] or 
    board[7] == board[4] == board[1] or 
    board[7] == board[4] == board[1] or 
    board[6] == board[3] == board[0] or  
    board[8] == board[4] == board[0] or
    board[6] == board[4] == board[2]):
        print(f'{active_player[0]} wins!')
        game_status = 0
    #checks the draw conditions
    if (board[0] != 1 and board[1] != 2 and board[2] != 3 and 
    board[3] != 4 and board[4] != 5 and board[5] != 6 and 
    board[6] != 7 and board[7] != 8 and board[8] != 9):
        print('stalemate!')
        game_status = 0
    #  changes it to the next player's turn
    if game_status == 1:
        turn_switch()
```

```python
def playagain():
    global board
    global player1
    global player2
    # asks the players if they would like to play again
    pag = input('would you like to play again? yes/no ')
    # if yes, resets the board
    if pag.lower() == 'yes':
        board = [1,2,3,4,5,6,7,8,9]
        player1 = ''
        player2 = ''
        game()
    # if no, ends game
    elif pag.lower() == 'no':
        print('Thank you for playing :)')
    #if input is invalid, runs the function again
    else:
        print('please input yes or no')
        playagain()
```

```python
def turn_switch():
    global turn
    global active_player
    if turn == 1:
        active_player = ['player 1', player1]
        turn = 0
    elif turn == 0:
        active_player = ['player 2', player2]
        turn = 1
    player_turn()
```
